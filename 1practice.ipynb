```python
# Data Understanding
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

```python
# df_2015= pd.read_csv('2015.csv')
# df_2016= pd.read_csv('2016.csv')
# df_2017= pd.read_csv('2017.csv')
# df_2018= pd.read_csv('2018.csv')
df_2019 = pd.read_csv('2019.csv')

# print(df_2015.head())
# print(df_2016.head())
# print(df_2017.head())
# print(df_2018.head())
print(df_2019.head())
```

```python
# Data Exploration
df_2019.info()
# df_2019.isnull().sum()
# df_2019.duplicated().sum()
# df_2019.describe()
```

```python
# Correlation
plt.figure(figsize=(10,8))
sns.heatmap(df_2019.corr(numeric_only=True), annot= True, cmap= 'coolwarm') #The numeric_only parameter is set to True to include only numerical columns in the correlation calculation.
plt.title('Correlation Matrix')
plt.show()
```

```python
# Barplot/graph
top_10_highest= df_2019.sort_values(by="Score", ascending=False).head(10)
plt.figure(figsize=(10,8))
sns.barplot(x="Score", y="Country or region", data=top_10_highest)
plt.title("Top 10 Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.show()
```

```python
# Box plot
plt.figure(figsize= (8,6))
sns.boxplot(y="Healthy life expectancy", data= df_2019)
plt.title('Distribution of HLE across countries')
plt.ylabel('HLE')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(x="GDP per capita", y="Score", data=df_2019)
plt.title("GDP per capita vs Happiness Score")
plt.xlabel("GDP per capita")
plt.ylabel("Happiness Score")
plt.show
```

```python
# Pair plot
sns.pairplot(df_2019[['GDP per capita','Social support','Healthy life expectancy','Score']])
plt.suptitle("Pair plot of key features", y=1.02)
plt.show()
```
